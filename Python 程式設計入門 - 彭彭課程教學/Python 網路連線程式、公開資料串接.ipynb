```python
import urllib.request as request
import json

src="https://data.ntpc.gov.tw/api/v1/rest/datastore/382000000A-002800-001"
with request.urlopen(src) as response:
    data=json.load(response) 
# print(data)

# 將蒐集的數據名稱列表出來
clist=data["result"]["records"]
for air in clist:
    print(air["ItemName"])
    
# 將蒐集的數據名稱儲存
clist=data["result"]["records"]
with open("偵測氣體.txt", "w" ,encoding="utf-8") as file:
    for air in clist:
        file.write(air["ItemName"]+"\n")
```
