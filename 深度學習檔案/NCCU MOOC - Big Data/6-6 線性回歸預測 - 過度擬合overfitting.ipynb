```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from ipywidgets import interact
```

```python
x = np.linspace(0,5,50)  # 代表0到5之間，取50個點
y = 1.2*x +0.8 + 0.6*np.random.randn(50)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,1.2*x+0.8,"r")
```

```python
X=np.linspace(0,5,1000)

def my_fit(n):
    Y = 4*np.sin(n*X) +4
    plt.scatter(x,y)
    plt.plot(X,Y,"r")
    plt.show()
```

```python
my_fit(5)
```

```python
interact(my_fit,n=(1,500))
```

overfitting: 把過去的資料學得太像，精確到一模一樣。未來的資料因為沒背過，所以會不準確，隨意吐個答案出來
